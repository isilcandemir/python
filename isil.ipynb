# Library Importing

Import necessary libraries for the project.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Data reading

Now read the data from excels

```python
#Read data to dataframes
player_df = pd.read_excel('train_data.xlsx', sheet_name='Player')
point_df = pd.read_excel('train_data.xlsx', sheet_name='Points')
player_df
```

### Data Formatting

Reshape the data into a proper format for plotting

```python
ages_and_values = player_df.groupby('Club').mean()[['Market Value']]
club_vs_points = point_df.groupby('Club').mean()['Points']

plt.plot(club_vs_points.values)
plt.xticks(np.arange(0,len(club_vs_points.values)),list((club_vs_points.index)), rotation='vertical',ha='right')
plt.show()
plt.scatter(np.arange(0,len(club_vs_points.values)),club_vs_points.values)
#plt.xlabel(list((club_vs_points.index)))
plt.xticks(np.arange(0,len(club_vs_points.values)),list((club_vs_points.index)), rotation='vertical',ha='right')
plt.show()
```
